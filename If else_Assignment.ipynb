```python
#Question 7
# If all subjects > 80 → Science Stream
# If English > 80 and Math & Science > 50 → Commerce Stream
# If English > 80 and Social Studies > 80 → Humanities

eng = int(input("Enter marks in English: "))
math = int(input("Enter marks in Math: "))
sci = int(input("Enter marks in Science: "))
sst = int(input("Enter marks in Social Studies: "))

if eng > 80 and math > 80 and sci > 80 and sst > 80:
    print("Science Stream")
elif eng > 80 and math > 50 and sci > 50:
    print("Commerce Stream")
elif eng > 80 and sst > 80:
    print("Humanities")
else:
    print("No stream will be allotted")
```

```python
#Question 6
#If Marked Price > 10000 → 20% discount
#If Marked Price > 7000 and ≤ 10000 → 15% discount
#If Marked Price ≤ 7000 → 10% discount
mp = float(input("Enter the Marked Price: "))

if mp > 10000:
    discount = mp * 0.20
elif mp > 7000 and mp <= 10000:
    discount = mp * 0.15
else:
    discount = mp * 0.10

net_amount = mp - discount

print("Discount:", discount)
print("No discount:", net_amount)
```

```python
#Question 4
#If service more than 10 years → Bonus = 10% of salary
#If service between 6 and 10 years → Bonus = 8% of salary
#If service less than 6 years → Bonus = 5% of salary

salary = float(input("Enter your salary: "))
years = int(input("Enter years of service: "))

if years > 10:
    bonus = salary * 0.10
elif years >= 6 and years <= 10:
    bonus = salary * 0.08
else:
    bonus = salary * 0.05

print("Net Bonus Amount =", bonus)
```

```python
#Question 3
# Accept the age of 4 people as input
age1 = int(input("Enter age for Person 1: "))
age2 = int(input("Enter age for Person 2: "))
age3 = int(input("Enter age for Person 3: "))
age4 = int(input("Enter age for Person 4: "))

# Assume the first person is the youngest initially
youngest = age1

# Use if-elif statements to find the true youngest age
if age2 < youngest:
    youngest = age2
elif age3 < youngest:
    youngest = age3
elif age4 < youngest:
    youngest = age4

# Display the youngest age
print(f"The youngest person's age is: {youngest}")
```

```python
#Question 2
#> 90 → Grade A
#> 80 and ≤ 90 → Grade B
#≥ 60 and ≤ 80 → Grade C
#Below 60 → Grade D

per = float(input("Enter percentage: "))

if per > 90:
    print("Grade A")
elif per > 80 and per <= 90:
    print("Grade B")
elif per >= 60 and per <= 80:
    print("Grade C")
else:
    print("Grade D")
```

```python
#Question 1
#First 100 units → No charge
#101 – 200 units → ₹5 per unit
#Above 200 units → ₹10 per unit
units = int(input("Enter number of units: "))
bill = 0

if units > 100:
    bill = bill + (min(units, 200) - 100) * 5

if units > 200:
    bill = bill + (units - 200) * 10

print("Total Electricity Bill =", bill)
```
